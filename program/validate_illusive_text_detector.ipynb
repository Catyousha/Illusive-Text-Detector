```python
import os
VAL_DIR = "../human-test-data"
```

```python
def generate_val_schema():
    """
    [
        {
            'document_file': 'doc_1.pdf',
            'text_file': 'text_1.txt'
        }
    ]
    """
    test_schema = []
    for root, dirs, files in os.walk(VAL_DIR):

        document_files = [f for f in files if f.endswith('.pdf')]
        actual_files = [f for f in files if f.endswith('.txt')]
        assert(len(document_files) == len(actual_files))

        test_schema = [
            {
                "document_file": document_files[i],
                "text_file": actual_files[i]
            } for i in range(0, len(document_files))
        ]
    return test_schema
```

```python
val_schema = generate_val_schema()
a = open(f"{VAL_DIR}/{val_schema[0]['text_file']}")
a.read()
a.close()
```

```python
from illusive_text_detector import IllusiveTextDetector
import pandas as pd
pd.set_option('max_colwidth', 400)
def validate_detector():
    detector = IllusiveTextDetector()
    val_schema = generate_val_schema()

    document_name = []
    predicted_chars = []
    actual_chars = []
    passed = []

    for idx, val in enumerate(val_schema):
        
        val_pdf_loc = f"{VAL_DIR}/{val['document_file']}"
        val_txt_loc = f"{VAL_DIR}/{val['text_file']}"

        txt_file = open(val_txt_loc)
        actual = txt_file.read()
        txt_file.close()

        predicted = detector.detect(val_pdf_loc)['chars_data']
        predicted = [x['char'] for x in predicted if x['label'] == 'ILLUSIVE' and x['char'] != ' ']
        actual = actual.split()

        document_name.append(f"Document {idx+1}")
        predicted_chars.append(predicted)
        actual_chars.append(actual)
        passed.append(predicted == actual)
    
    result = {
        "Document": document_name,
        "Actual Characters": actual_chars,
        "Predicted Characters": predicted_chars,
        "Passed": passed,
    }

    df = pd.DataFrame.from_dict(result)
    display(df)
validate_detector()
```
